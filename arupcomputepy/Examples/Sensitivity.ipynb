```python
import ArupCompute
import pandas as pd
from collections import OrderedDict
import ipywidgets as widgets
```

```python
library = 'designcheck'
calculation = 'mep/mechanical/psychrometrics/humidair'
```

```python
variables = {
    'T_db': [5,10,15,20],
    'InputType': ['RH'],
    'InputValue' : range(10, 41, 10),
    'Plot' : ['False']
}
```

```python
import itertools
def run_sensitivity(variables, library, calculation, limit=True):
    l = widgets.Label("Preparing calculations")
    display(l)
    calc_list = []
    for item in list(itertools.product(*variables.values())):
        params = dict(zip(variables.keys(), item))
        calc_list.append(ArupCompute.PrepareInputs(library, calculation, params))
    
    if limit and len(calc_list) > 50:
        raise Exception("{n} is too many calculations".format(n=len(calc_list)))
    
    l.value="Running calculations..."
    responses =  ArupCompute.ExecuteCalculationsSync(calc_list, useArupProxy=False)
    l.value="Done."
    
    return responses
```

```python
results_dict = OrderedDict()
responses = run_sensitivity(variables, library=library, calculation=calculation)
for k in responses[0]['results'].keys():
    fieldname = responses[0]['results'][k]['description']
    results_dict[fieldname] = list(r['results'][k]['value'] for r in responses)

df = pd.DataFrame(results_dict)
display(df)
```
